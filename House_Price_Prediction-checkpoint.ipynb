```python
def plt_scatter_chart(df, location):
    # Filter the DataFrame based on location
    filtered_df = df[df['location'] == location]
    
    # Create a scatter plot (for example, 'price' vs. 'area')
    plt.figure(figsize=(10, 6))
    plt.scatter(filtered_df['Area Size'], filtered_df['price'], color='blue')
    
    plt.title(f'Scatter Plot for {location}')
    plt.xlabel('Area Size')
    plt.ylabel('Price')
    plt.grid(True)
    plt.show()

# Example usage
plt_scatter_chart(df, "Gulshan-e-Iqbal Town")
```

```python
# Import the necessary libraries
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error

# Load your dataset
# Assuming 'df' is your DataFrame containing the dataset

# Check for NaN values and replace or drop them
print("Checking for NaN values...")
print(df.isna().sum())  # Prints the number of NaN values per column

# Option 1: Drop rows with NaN values (use with caution)
df = df.dropna()

# Option 2: Fill NaN values with some strategy
# df = df.fillna(df.median())  # Filling NaNs with the median value of each column

# Check for Infinity values and replace them
print("Replacing Infinity values...")
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()

# Ensure no extremely large values (outliers) exist
print("Checking for excessively large values...")
print(df.describe())  # Check the statistical summary to identify any potential outliers

# Define the features and target
X = df.drop(['price'], axis=1)
Y = df['price']

# Identify the categorical features
categorical_features = ['location']  # Add other categorical features if any

# Create a ColumnTransformer to handle the encoding
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'  # Leave other columns untouched
)

# Create a pipeline that includes the preprocessor and the regressor
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', DecisionTreeRegressor(criterion='friedman_mse', splitter='random', random_state=0))
])

# Proceed with the train-test split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)

# Fit the model
pipeline.fit(X_train, y_train)

# Predict on the test data
y_pred = pipeline.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
score = pipeline.score(X_test, y_test)

print(f"Model Score: {score}")
print(f"Mean Squared Error: {mse}")
```
